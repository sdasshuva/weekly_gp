```python
from datetime import date
from datetime import timedelta
today = date.today() # or you can do today = date.today() for today's date

for i in range(1,8):
    excel_date = today - timedelta(days=i)
    print(excel_date)
#     stored_file_location = "/root/PRTG_DAILY_REPORT/weekly_report_gp/"
    excel_name_old = str(excel_date) + "-18-00-00-all-site-report-prtg-dhaka-peak.xlsx"
    
#     # check eligibility
    if i == 1:
        dat1 =  pd.read_excel(excel_name_old, index_col=0)  
        dat1["Eligible"] = "NO"
        excel_name_last =  str(excel_date) + "-18-00-00-all-site-report-prtg-dhaka-peak.xlsx"
        excel_name_pd_last = pd.read_excel(excel_name_last, index_col=0)
        dat1["count"] = 0
        
        dat1 = dat1.set_index('site_name')
        dat1 = dat1.loc[~dat1.index.duplicated(keep='first')]
        
        
        
        
        
    else:
        
        excel_name_pd = pd.read_excel(excel_name_old, index_col=0)
        excel_name_pd = excel_name_pd.drop(columns=["capacity_Mbps","count","Eligible","Starting Date","Peak"])
        excel_name_pd = excel_name_pd.set_index('site_name')
        excel_name_pd = excel_name_pd.loc[~excel_name_pd.index.duplicated(keep='first')]
        
        dat1 = pd.concat([dat1, excel_name_pd], axis=1, join="inner")
#         result = pd.concat([df1, df2], axis=1, join="inner")
```
